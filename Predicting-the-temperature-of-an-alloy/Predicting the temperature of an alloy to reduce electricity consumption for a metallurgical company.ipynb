**Описание задачи и технологического процесса.**

Чтобы оптимизировать производственные расходы, металлургический комбинат «Стальная птица» решил уменьшить потребление электроэнергии на этапе обработки стали. Для этого комбинату нужно контролировать температуру сплава. Задача — построить модель, которая будет использоваться для имитации технологического процесса.

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

Из сплава выводится сера (этот процесс — десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (от англ. tribe — «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (от англ. slab — «плита»).

# Предобработка данных.

## Стандартизация названий столбцов.

```python
data_arc = data_arc.rename(columns={
    'Начало нагрева дугой': 'time_start',
    'Конец нагрева дугой': 'time_end',
    'Активная мощность': 'power_active',
    'Реактивная мощность': 'power_reactive'
})

data_bulk.columns = (
        data_bulk.columns
        .str.replace(pat='B', repl='b')
        .str.replace(pat=' ', repl='_')
)
    
data_wire.columns = (
        data_wire.columns
        .str.replace(pat='W', repl='w')
        .str.replace(pat=' ', repl='_')
)
    
data_gas = data_gas.rename(columns={
    'Газ 1': 'gas_1'
})

data_temp = data_temp.rename(columns={
    'Время замера': 'time',
    'Температура': 'temperature'
})
```

## Обработка пропущенных значений.

```python
# в таблицах проволочных и сыпучих материалов заменим
# пропущенные значения на 0, т.к. материал не добавлялся
data_bulk = data_bulk.fillna(0)
data_wire = data_wire.fillna(0)
display(data_bulk.head())
data_wire.tail()
```

```python
# значения таблицы температур являются ключевыми для задачи
# удалим те значения, которые не содержат этих данных
data_temp = data_temp.dropna().reset_index(drop=True)

# для некоторых партий осталось только одно измерение
display(data_temp.head())
data_temp.tail()
```

```python
# партии с единственным измерением не подходят
# для имитации технологического процесса, поэтому
# такие партии необходимо удалить
for key in data_temp.key.unique():
    count = (data_temp.key == key).sum()
    if count < 2:
        data_temp = (
            data_temp[data_temp.key != key]
            .reset_index(drop=True)
        )
display(data_temp.head())
data_temp.tail()        
```

## Обработка аномальных значений.

```python
# найдем партии с аномальной температурой
low_temp_keys = list(data_temp[data_temp.temperature < 1500].key)

# найдем партии с аномальной мощностью
neg_power_keys = list((data_arc[(data_arc.power_active < 0) | (data_arc.power_reactive < 0)]).key)

abnormal_keys = low_temp_keys + neg_power_keys

# создадим функцию, удаляющие партии с аномалиями
def delete_abnormal(data):
    return data.loc[~data.key.isin(abnormal_keys)]

# удалим аномалии из данных
data_arc = delete_abnormal(data_arc)
data_bulk = delete_abnormal(data_bulk)
data_wire = delete_abnormal(data_wire)
data_gas = delete_abnormal(data_gas)
data_temp = delete_abnormal(data_temp)
```

```python
def show_distribution_density(data, column_name_1, column_name_2, x_label):
    for sample, sample_name, color in zip([data[column_name_1], data[column_name_2]],
                                          [column_name_1, column_name_2],
                                          ['red', 'blue']):
        sns.histplot(data=sample,
                            bins=100,
                            kde=True,
                            stat='density',
                            color=color,
                            label=sample_name).set_title('Плотность распределения:' +
                                                         f' {column_name_1} и {column_name_2}')
        sns.set(rc={'figure.figsize':(15,10)})
        plt.axvline(x=sample.mean(), ls='--', lw=3, color=color)
        plt.legend()
        #plt.xlim(xmin=0, xmax = 150)
        plt.xlabel(x_label)
        plt.ylabel('Плотность распределения');
```

```python
show_distribution_density(data_arc, 'power_active', 'power_reactive', 'Мощность')
```

2. Однако есть и `сходства`: самые часто встречающиеся значения обоих распределений меньше среднего среднего значения.
3. Отличия средних значений мощностей - чуть меньше 0.2 в пользу активной мощности.

```python
# перевод дат в тип 'datetime'
def change_type(data, column):
    data[column] = pd.to_datetime(data[column], format='%Y-%m-%d %H:%M:%S')
    return data

data_arc = change_type(data_arc, 'time_start')
data_arc = change_type(data_arc, 'time_end')
data_temp = change_type(data_temp, 'time')
```

## Группировка и агрегация таблиц по партиям.

### `data_arc`.

```python
# найдем время нагревания в секундах для каждой итерации 
data_arc['time_of_heating'] = (
    (data_arc.time_end - data_arc.time_start) /\
    np.timedelta64(1, 's')
)

# создадим сводную таблицу для данных об электродах,
# в которую войдут суммарное время нагревания, а также
# средние мощности для всего времени нагревания
data_arc = pd.pivot_table(data_arc,
                          index='key',
                          values=['time_of_heating',
                                  'power_active',
                                  'power_reactive'],
                   aggfunc={'time_of_heating': 'sum',
                            'power_active': 'mean',
                            'power_reactive': 'mean'}).reset_index()

display(data_arc.head())
data_arc.tail()
```

### `data_temp`.

```python
# создадим сводную таблицу для данных о температуре,
# в которую войдут минимальные и максимальные
# значения времени и температуры
data_temp = (
    pd.pivot_table(data_temp,
                   index='key',
                   values=['time',
                           'temperature'],
                   aggfunc={'time': ['first', 'last'],
                            'temperature': ['first', 'last']})
    .reset_index()
)
data_temp.tail()
```

### Добавление новых признаков.

#### Полная мощность и коэффициент мощности.

Показатели мощности напрямую влияют на изменение температуры, а их отношение можно отобразить следующей схемой:


Зная активную мощность $P$ и реактивную мощность $Q$, можно добавить в таблицу новые признаки: полную мощность $S$ и коэффициент мощности $P$.

$$
S = \sqrt{P^{2} + Q^{2}}
$$

$$
PF = \dfrac{P}{S}
$$

```python
# найдем полную мощность
data_merged['power_full'] = (
    np.sqrt(
        np.square(data_merged.power_active) + \
        np.square(data_merged.power_reactive)
    )
)

# найдем коэффициент мощности
data_merged['power_coeff'] = data_merged.power_active / data_merged.power_full
```

#### Время нагревания и общее время процесса.

Время, во время которого нагревают сплав, напрямую влияет на изменение температуры. Зная время нагревания $t_1$ и общее время $t_{sum}$, можно найти долю времени нагревания $t_{ratio}$ и остальное время $t_2$.

$$
t_{ratio} = \dfrac{t_1}{t_{sum}}
$$

$$
t_2 = t_{sum} - t_1
$$

```python
# добавим в таблицу отношение времени нагревания
# и общего времени процесса, которое будет показывать,
# какую долю времени занимает нагревание сплава
data_merged['time_ratio'] = (
    data_merged.time_of_heating / \
    data_merged.time_of_process
)

# также добавим время, когда сплав не нагревался
data_merged['time_other'] = (
    data_merged.time_of_process - \
    data_merged.time_of_heating
)
```

#### Суммарный объем добавляемых материалов.

```python
# добавим суммарный объем добавляемых материалов
data_merged['bulk_total'] = data_merged[['bulk_1',
                                         'bulk_2',
                                         'bulk_3',
                                         'bulk_4',
                                         'bulk_5',
                                         'bulk_6',
                                         'bulk_7',
                                         'bulk_8',
                                         'bulk_9',
                                         'bulk_10',
                                         'bulk_11',
                                         'bulk_12',
                                         'bulk_13',
                                         'bulk_14',
                                         'bulk_15']].sum(axis=1)

data_merged['wire_total'] = data_merged[['wire_1',
                                         'wire_2',
                                         'wire_3',
                                         'wire_4',
                                         'wire_5',
                                         'wire_6',
                                         'wire_7',
                                         'wire_8',
                                         'wire_9']].sum(axis=1)

data_merged['bulk_wire_total'] = data_merged[['bulk_total',
                                              'wire_total']].sum(axis=1)
```

# Отчет о проекте и разработка рекомендаций для заказчика.

1. **Введение.** Для оптимизации производственных расходов металлургический комбинат «Стальная птица» решил уменьшить потребление электроэнергии на этапе обработки стали. Для этого комбинату нужно контролировать температуру сплава. В данный момент это происходит с помощью остановки процесса и измерения температуры для каждой итерации. *Заказчик хочет использовать машинное обучение для предсказания точной конечной температуры, что позволит уменьшить электроэнергию и, вероятно, количество добавляемых в сплав материалов.*

    
2. **Технологический процесс.** Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. Из сплава выводится сера (этот процесс — десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат. Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки. Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов.


3. **Данные.** *Заказчиком предоставлены 7 таблиц и 48 627 строк данных для 3241 партии стали за период с 2019-05-03 по 2019-09-06*. Для дальнейшего обучения моделей данные нуждались в обработке. Во-первых, были обработаны пропуски: пропущенные значения объема материалов были заполнены нулями, а строки, в которых были пропущена информация о температуре, удалены. После требовалось убрать такие партии, где температура измерялась меньше 2 раз. Во-вторых, из данных удалены партии с аномальными значениями: температура сплава менее 1500 градусов и мощность менее 0. В-третьих, данные были объединены в единую таблицу по ключу *key* с частью важных признаков: суммарное время нагревания, средняя активная и реактивная мощности для общего времени нагревания, суммарное время процесса с начальным и конечным значением температуры. В единую таблицу также были добавлены новые признаки: полная мощность, коэффициент мощности, доля времени нагревания, время без нагревания, суммарный объем сыпучих материалов, суммарный объем проволочных материалов, суммарный объем всех добавленных материалов.


4. **Исследовательский анализ.** Были изучены скрытые взаимосвязи в данных, а также исключены второстепенные с точки зрения будущего предсказания данные. С одной стороны, был определено, что между конечной температурой сплава и временем его нагревания есть слабая корреляционная связь: чем дольше нагревают сплав, тем больше увеличивается температура или меньше снижается. Между конечной температурой сплава и суммарным объемом добавляемых материалов есть слабая корреляционная связь: чем больше добавляют материала в сплав, тем выше конечная температура. Это говорит о том, что время нагревания и объем добавляемых материалов помогут в обучении модели. С другой стороны, между конечной температурой сплава и активной мощностью нет корреляционной связи. Таким образом доказано, что признаки мощности не помогут модели предсказать точную конечную температуру, поэтому были удалены из данных. Также был проведен анализ мультиколлинеарности признаков. Он показал, что необходимости удалять какие-либо другие признаки из данных нет.


5. **Машинное обучение.** Данные были разделены случайным образом: 75% тренировочных данных и 25% тестовых данных. Признаки состояли только из численных значений. Далее были созданы 2 пайплайна, которые включали себя последовательные шаги: масштабирование признаков методом стандартизации, обучение модели с подбором оптимальных гиперпараметров, кросс-валидация. Моделями для пайплайна стали: *линейная регрессия* и *CatBoostRegressor*.


6. **Результаты.** Лучшая модель на этапе кросс-валидации тренировочных данных - модель линейной регрессии. Метрика MAE составила 6.18. Модель показала значительно лучший результат, если бы мы предсказывали конечную температуру с помощью среднего значения - средняя абсолютная ошибка константной модели составила 8.28. Модель линейной регресси с лучшими гиперпараметрами была протестирована на тестовых данных. MAE лучшей модели составил 5.78.


7. **Анализ важности признаков.** Были изучен влияние каждого признака в предсказание конечной температуры. *Признаками с наибольшим вкладом стали суммарные объемы сыпучих и проволочных материалов, а также 2 проволочных материала (wire_1 и wire_2).* Анализ важности признаков показал, что ни начальная температура, ни мощность, ни время нагревания не вляют так сильно, как объем добавляемых сыпучих и проволочных материалов.


8. **Перспективы.** Дальнейшая работа в этом направлении может существенно улучшить как предсказание конечной температуры, так и экономию средств на электроэнергии. Во-первых, можно увеличить количество признаков для обучения модели: физические и химические свойства добавляемых материалов, их вес, а также свойства первоначального сплава. Во-вторых, сама модель машинного обучения может быть расширена посредством использования таких ансамблевых методов, как стекинг или бэггинг, что позволит предсказывать результат с помощью целого каскада моделей. В-третьих, в дальнейшем можно предсказывать не только конечную температуру, но и общую стоимость процесса, что позволит изучить влияние каждого этапа и компонента на экономику процесса напрямую.
