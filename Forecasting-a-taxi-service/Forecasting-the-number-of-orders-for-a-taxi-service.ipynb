# Прогнозирование количества заказов для сервиса такси

<div style="border: dashed 10px; background-color: #41778C; color:#ffffff; padding: 20px">

Задача. Сервис такси собрал исторические данные о заказах в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Необходимо построить модель для такого предсказания. Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

## Подготовка выборок

```python
df.head(10)
```

```python
# для последующего обучения снова возьмем неизмененный датафрейм
df = pd.read_csv('taxi.csv', index_col=[0], parse_dates=[0])
df.sort_index(inplace=True)
df = df.resample('1H').sum()
```

<div style="background-color: #74D496; color:#ffffff; padding: 15px; border-radius: 15px;">

### Создание признаков

```python
step = 168

# добавим календарные признаки в соответствии с выявленной сезонностью
df['hour'] = df.index.hour
df['dayofweek'] = df.index.dayofweek
    
# добавим признак "отстающих значений" с отступом 168 строк
# + 1 строка, чтобы вычисление скользящего среднего не включало в себя
# текущее значение ряда
for lag in range(1, step + 1):
    df['lag_{}'.format(lag)] = df['num_orders'].shift(lag)
        
# добавим признак "скользящего среднего" с размером окна, равным 1 неделе
df['rolling_mean'] = df['num_orders'].shift().rolling(step).mean()
```

```python
df.shape
```

<div style="background-color: #74D496; color:#ffffff; padding: 15px; border-radius: 15px;">

### Создание выборок

```python
df = df.dropna()
train, test = train_test_split(df, test_size=0.2, shuffle=False)

# убедимся, что данные разбиты по выборкам правильно
display(train.index.min(), train.index.max())
display(test.index.min(), test.index.max())
```

```python
features_train = train.drop('num_orders', axis=1)
target_train = train['num_orders']

features_test = test.drop('num_orders', axis=1)
target_test = test['num_orders']
```

<div style="background-color: #E87FAA; color:#ffffff; padding: 15px; border-radius: 15px;">

### Вывод

1. Подготовлены `новые признаки`: час в сутках, день недели, признаки "отстающих значений" (максимальное отставание - 168 часов) и "скользящее среднее" (окно - 168 часов). 


2. Подготовлены `обучающая и тестовая выборки` в отношении 4:1. Для сохранения временного ряда перемешивание данных не проводилось.

<div style="background-color: #74D496; color:#ffffff; padding: 15px; border-radius: 15px;">

## Создание пайплайнов

```python
# создадим объект для стандартизации количественных
# признаков методом масштабирования
column_transformer = make_column_transformer((StandardScaler(), make_column_selector(dtype_include='number')))
```

```python
# создадим пайплайн модели линейной регрессии
model_lr = LinearRegression()
pipeline_lr = make_pipeline(column_transformer, model_lr)
params_lr = {
    'linearregression__fit_intercept': [True, False]
}
```

```python
# создадим модель стекинга;
# используем две одинаковые базовые модели линейной регрессии
# и модель случайного леса для в качестве мета-модели
base_models = [('lin_reg_1', pipeline_lr),
               ('lin_reg_2', pipeline_lr)]

model_rfr = RandomForestRegressor(random_state=RS)

model_sr = StackingRegressor(base_models,
                             model_rfr)
```

```python
# создадим пайплайн для модели случайного леса
pipeline_rfr = make_pipeline(column_transformer, model_rfr)
params_rfr = {
    'randomforestregressor__max_depth': list(np.arange(2, 5, step=1)) + [None],
    'randomforestregressor__n_estimators': list(np.arange(2, 8, step=1))
}
```

```python
# создадим пайплайн модели CatBoostRegressor
model_cbr = CatBoostRegressor(random_state=RS,
                              verbose=False)
pipeline_cbr = make_pipeline(column_transformer, model_cbr)
params_cbr={
    'catboostregressor__learning_rate': list(np.arange(0.01, 0.11, step=0.02)),
    'catboostregressor__iterations': list(np.arange(50, 151, step=50))
}
```

<div style="background-color: #00AC3A; color:#ffffff; padding: 30px; border-radius: 15px;">

## Обучение и валидация моделей

<div style="background-color: #74D496; color:#ffffff; padding: 15px; border-radius: 15px;">

### Линейная модель - `LinearRegression`

```python
# создадим объект TimeSeriesSplit для корректного
# разделения временного ряда при кросс-валидации
tscv = TimeSeriesSplit()

model_lr_grid = GridSearchCV(pipeline_lr,
                             params_lr,
                             cv=tscv,
                             scoring='neg_root_mean_squared_error')
model_lr_grid.fit(features_train, target_train);

rmse_lr = abs(model_lr_grid.best_score_)
rmse_lr
```

<div style="background-color: #74D496; color:#ffffff; padding: 15px; border-radius: 15px;">

### Стекинг - `StackingRegressor`

```python
scores_stacking = cross_val_score(model_sr,
                                  features_train,
                                  target_train,
                                  cv=tscv,
                                  scoring='neg_root_mean_squared_error')

rmse_sr = abs(sum(scores_stacking) / len(scores_stacking))
rmse_sr
```

<div style="background-color: #74D496; color:#ffffff; padding: 15px; border-radius: 15px;">

### Бэггинг - `RandomForestRegressor`

```python
model_rfr_grid = GridSearchCV(pipeline_rfr, 
                              params_rfr,
                              cv=tscv,
                              scoring='neg_root_mean_squared_error')
model_rfr_grid.fit(features_train, target_train);

rmse_rfr = abs(model_rfr_grid.best_score_)
rmse_rfr
```

<div style="background-color: #74D496; color:#ffffff; padding: 15px; border-radius: 15px;">

### Градиентный бустинг - `CatBoostRegressor`

```python
model_cbr_grid = GridSearchCV(pipeline_cbr,
                              params_cbr,
                              cv=tscv,
                              scoring='neg_root_mean_squared_error')
model_cbr_grid.fit(features_train, target_train);

rmse_cbr = abs(model_cbr_grid.best_score_)
rmse_cbr
```

<div style="background-color: #74D496; color:#ffffff; padding: 15px; border-radius: 15px;">

### Простейшие модели

Сравним предсказание полученных моделей с простым предсказанием признаков: моделью константного медианного значения и моделью, предсказанием которой будет предыдущее значение ряда.

<div style="border: solid green 3px; padding: 10px; border-radius: 10px">

#### Константная модель

```python
pred_median = np.ones(target_test.shape) * target_train.median()
rmse_const_median = mean_squared_error(target_test, pred_median, squared=False)
rmse_const_median
```

<div style="border: solid green 3px; padding: 10px; border-radius: 10px">

#### Модель с предыдущим значением ряда

```python
pred_previous = target_test.shift()
pred_previous.iloc[0] = target_train.iloc[-1]
rmse_previous = mean_squared_error(target_test, pred_previous, squared=False)
rmse_previous
```

<div style="background-color: #E87FAA; color:#ffffff; padding: 15px; border-radius: 15px;">

### Вывод

```python
pd.DataFrame({'Models' : ['LinearRegression',
                          'StackingRegressor',
                          'RandomForestRegressor',
                          'CatBoostRegressor',
                          'Model: constant median',
                          'Model: previous value'],
              'RMSE' : [rmse_lr, rmse_sr, rmse_rfr, rmse_cbr, rmse_const_median, rmse_previous]},
              index = [1, 2, 3, 4, 5, 6])
```

<div style="background-color: #E87FAA; color:#ffffff; padding: 15px; border-radius: 15px;">

1. Лучшая модель на этапе валидации - `CatBoostRegressor`, однако при учете времени обучения для тестирования выберем `LinearRegression`.
    
    
2. `Проверка на адеватность модели пройдена`: метрики всех моделей в 2 и более раз лучше, чем на простом предсказании медианным числом и предыдущем значении ряда.
